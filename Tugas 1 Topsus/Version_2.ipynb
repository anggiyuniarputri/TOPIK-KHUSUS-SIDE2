```python
import pandas
import os
import re
from nltk import pos_tag, word_tokenize
```

```python
data_set = pandas.read_csv(os.getcwd() +'/Dataset Canon S100 Reanotasi.csv', sep='delimiter', header=None)

list_pos_tag = []
i = 1
for data in data_set[0]:
    print(str(i) +' - ' +data)
    punctuated = re.sub("[^a-zA-Z]", " ", data)
    print(punctuated)
    words = word_tokenize(punctuated)
    print(words)
    postag = pos_tag(words)
    list_pos_tag.append(postag)
    print(postag)
    i += 1
    print('')
    
raw_data = {'pos_tag': list_pos_tag}
df = pandas.DataFrame(raw_data, columns=['pos_tag'])
df.to_csv(os.getcwd() +'/Version_2.csv', index=False)
```
